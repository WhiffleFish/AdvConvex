```python
import pandas as pd
import cvxpy as cvx
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
import scipy
from matplotlib.lines import Line2D
```

```python
N = 100
m = N
L = 5
h = [np.exp(-np.square(x)/2) for x in range(-2,3)]
x = np.zeros(N)
x[9] = 1
x[12] = -1
x[49] = 0.3
x[70] = -0.2
h
```

# 1

```python
def unit_basis(i, N):
    b = np.zeros(N)
    b[i] = 1
    return b

def blur(x):
    return convolve(x, h, mode='wrap')

def implicit2explicit(f, N):
    m = len(f(unit_basis(0,N)))
    B = np.zeros((m, N))
    for i in range(N):
        b = unit_basis(i,N)
        B[i,:] = f(b)
    return B
```

```python
B = implicit2explicit(blur, N)
all((B @ x) == blur(x)) # True
```

# 2

```python
σ = 0.02
ϵ = σ * np.sqrt(N)
y = B @ x + np.random.randn(N) * σ
X = cvx.Variable(N)
obj = cvx.Minimize(cvx.norm1(X))
prob = cvx.Problem(obj, [cvx.norm2(B@X - y)**2 <= ϵ**2])
prob.solve()
x̂ = X.value
```

```python
plt.plot(y, label="y", marker='s', markersize=4, fillstyle='none')
plt.plot(x, label="x", marker='x', markersize=4, fillstyle='none')
plt.plot(x̂, label="x̂", marker='o', markersize=4, fillstyle='none')
plt.legend()
```

# 3

```python
X2 = cvx.Variable(N)
λ = prob.constraints[0].dual_value
prob2 = cvx.Problem(
    cvx.Minimize(
        cvx.norm1(X2) + λ*cvx.norm2(B @ X2 - y)**2
    )
)
prob2.solve()
```

```python
np.linalg.norm(X2.value - X.value, 2)
```

```python
def F(x):
#     return np.fft.fft(x)
    return scipy.fft.fft(x)

def Ⅎ(x):
#     return np.fft.ifft(x)
    return scipy.fft.ifft(x)
    
def ℋ(z,h=h):
#     h_pad = np.pad(h, (0,len(z)-len(h)))
    h_pad = np.pad(h, (47,48))
    return F(h_pad) * z

def ℋ_star(z,h=h):
#     h_pad = np.pad(h, (0,len(z)-len(h)))
    h_pad = np.pad(h, (47,48))
    return np.conjugate(F(h_pad)) * z

def B_func(x):
    return Ⅎ(ℋ(F(x)))

def B_func_star(x):
    return Ⅎ(ℋ_star(F(x)))
```

```python
a = np.random.rand(100)*10
b = np.random.rand(100)*10
Bs = np.conj(B).T
d1 = np.vdot(B@a, b)
d2 = np.vdot(a, Bs@b)
print(d1)
print(d2)
```
